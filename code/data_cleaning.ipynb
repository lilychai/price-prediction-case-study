```python
import pandas as pd
import numpy as np
import zipfile
```

```python
# df = pd.read_csv(zipfile.ZipFile('data/Train.zip', mode='r').open('Train.csv'), low_memory=False)
df = pd.read_csv('data/Train.csv', low_memory=False, parse_dates=['saledate'])
```

```python
def cleaning(df):
    """
    :type df: pandas.DataFrame
    :return: pandas.DataFrame
    """
    
    for cname in df.columns[-13:]:
        df[cname] = pd.Categorical(df[cname]).codes
        
    return df
```

```python
df2 = cleaning(df)
```
